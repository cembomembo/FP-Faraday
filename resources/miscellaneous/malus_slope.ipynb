```python
import numpy as np
import matplotlib.pyplot as plt

# Malus' law (not normalized): I(phi) = I0 cos^2(phi)
phi_deg = np.linspace(0, 90, 600)   # degrees
I0 = 1.0                            # choose a convenient scale for plotting
I = I0 * np.cos(np.deg2rad(phi_deg))**2

# Working point at 45°
phi0_deg = 45.0
I_wp = I0 * np.cos(np.deg2rad(phi0_deg))**2

# Derivative: d/dphi [I0 cos^2(phi)] = -I0 sin(2phi), with phi in radians
slope_per_rad = -I0 * np.sin(2 * np.deg2rad(phi0_deg))  # = -I0 at 45°

# Tangent line around 45°
phi_tan_deg = np.linspace(20, 70, 250)
I_tan = I_wp + slope_per_rad * np.deg2rad(phi_tan_deg - phi0_deg)

plt.figure(figsize=(7.0, 4.2))
plt.plot(phi_deg, I, label=r"$I(\phi) = I_0 \, \cos^2(\phi)$")
plt.plot(phi_tan_deg, I_tan, "--", label=r"Tangent at $\phi=45^\circ$")

# Mark and annotate working point
plt.scatter([phi0_deg], [I_wp], zorder=5)
plt.annotate(
    r"$\phi=45^\circ$" "\n" r"max slope",
    xy=(phi0_deg, I_wp),
    xytext=(57, 0.82 * I0),
    arrowprops=dict(arrowstyle="->"),
    fontsize=10
)

plt.xlim(0, 90)
plt.ylim(-0.05 * I0, 1.05 * I0)
plt.xlabel(r"Analyzer angle $\phi$ (degrees)")
plt.ylabel(r"Intensity $I$")

# Only show y-ticks at 0 and I0
plt.yticks([0, I0], [r"$0$", r"$I_0$"])

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
